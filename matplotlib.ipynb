```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,20)
x
y=x*x
z=x+y
plt.plot(x,y,x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.plot(x,z)
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
plt.tight_layout()
```

```python
 fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,color='red',marker='o',markersize='5')
axes[1].plot(x,z)
axes[0].set_xlabel("x-axis")
axes[0].set_ylabel("x-axis")
axes[1].set_xlabel("x-axis")
axes[1].set_ylabel("x-axis")
axes[1].set_xlimit([0,5])

plt.tight_layout()
```
